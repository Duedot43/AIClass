# The Question
What is the most popular form of serving (Box, bowl, cup, etc) ramen and what is that in each country

```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt

def isset(testVar, varName):
    try:
        test = testVar[varName]
        return True
    except:
        return False

data = pd.read_csv("ramen.csv")

country = pd.unique(data['Country'])

amount = data.count()['Brand']

count = {}

for x in range(amount):
    calculate = True
    try:
        stars = float(data.loc[x]['Stars'])
    except:
        calculate = False
    if calculate:
        if isset(count, data.loc[x]['Country']) == False:
            count[data.loc[x]['Country']] = {data.loc[x]['Style']: [stars, 1]}
        elif isset(count[data.loc[x]['Country']], data.loc[x]['Style']) == False:
            count[data.loc[x]['Country']][data.loc[x]['Style']] =  [stars, 1]
        else:
            count[data.loc[x]['Country']][data.loc[x]['Style']][0] += stars
            count[data.loc[x]['Country']][data.loc[x]['Style']][1] += 1
# 38 countrys

for key in count:
    key1 = key
    for key in count[key1]:
        count[key1][key] = count[key1][key][0]/count[key1][key][1]
        

thing = []
other = -1
for key in count:
    key1 = key
    other += 1
    if other <=5:
        for key in count[key1]:
            thing.append([key1, key, count[key1][key]])

thing

df = pd.DataFrame(thing)
#df = df.columns.values[0] = "Country"
#df.rename(index={0, 'c'})
#df_melt = pd.melt(df, id_vars=None, value_vars=['Cup', 'Tray', 'Pack', 'Bowl', 'Box', None, 'Can', 'Bar'])


popular = {}
other = -1
for key in count:
    key1 = key
    maximum = 0
    other += 1
    if other <= 5:
        for key in count[key1]:
            if count[key1][key] > maximum:
                maximum = count[key1][key]
                popular[key1] = [maximum, key]
popular
```
