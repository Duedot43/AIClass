```python
import pandas as pd
import seaborn as sb

data = pd.read_csv("spotify2024.csv",  encoding='unicode_escape')

months = {}

for x in data["Release Date"].tolist():
    x = x.split("/")
    if x[2] == "2024":
        try:
            months[int(x[0])] += 1
        except:
            months[int(x[0])] = 1

monthsNew = {"Months": {}, "Count": {}}
monthName = ["NaN", "Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
for x in range(1,13):
    try:
        monthsNew["Months"][x] = monthName[x]
        monthsNew["Count"][x] = months[x]
    except:
        monthsNew["Count"][x] = 0
print(monthsNew)


cleanExp = [0, 0]
for x in data["Explicit Track"]:
    cleanExp[x] += 1

ammount = cleanExp[0]+cleanExp[1]
cleanExp[0] = int((cleanExp[0]/ammount)*100)
cleanExp[1] = int((cleanExp[1]/ammount)*100)
print(cleanExp)

sb.barplot(data=monthsNew, x="Months", y="Count")
```

```python
import pandas as pd
import seaborn as sb

data = pd.read_csv("spotify2024.csv",  encoding='unicode_escape')

minimum = min(data['Track Score'])
maximum = max(data['Track Score'])

history = {}

for x in data['Track Score']:
    try:
        history[int(x)] += 1
    except:
        try:
            history[int(x)] = 1
        except:
            continue

for x in range(len(history)-1):
    try:
        if int(history[x]) <= 50:
            history.pop(x)
    except:
        print("EXP ", x )

sb.barplot(history)
history
```

```python
import pandas as pd
import seaborn as sb
import numpy as np

data = pd.read_csv("spotify2023.csv",  encoding='unicode_escape')

streams = []

for x in data['streams']:
    try:
        streams.append(int(x))
    except:
        print("error")

sb.boxplot(streams, orient="h")

summary = [int(np.min(streams)), int(np.percentile(streams, 25)), int(np.median(streams)), int(np.percentile(streams, 75)), int(np.max(streams))]
print(summary)
```
